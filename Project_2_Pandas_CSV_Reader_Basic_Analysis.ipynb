```python
# Filter rows, select columns & slice subsets
# Practicing this by finding outliers using IQR method

outliers = pd.DataFrame()
for i in df.select_dtypes(include=['float64', 'int64']).columns:
    iqr = df[i].quantile(0.75) - df[i].quantile(0.25)
    lower_bound = df[i].quantile(0.25) - (1.5 * iqr)
    upper_bound = df[i].quantile(0.75) + (1.5 * iqr)
    outliers = pd.concat([outliers, df[(df[i] < lower_bound) | (df[i] > upper_bound)]])
outliers
```

```python
# Save the outliers to a new CSV file
outliers.to_csv('outliers.csv', index=False)
```
